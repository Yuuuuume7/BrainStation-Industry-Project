## Statistic

```python
from scipy import stats
```

Hypothesis Test:

<center> $H_0$: $\mu_{Mentioned Keyword} = \mu_{No Keyword} $ $\;\; vs \;\;$  $H_1$: $\mu_{Mentioned Keyword} \neq \mu_{No keyword} $ </center>

```python
mentioned_keyword = np.array(keyword_df["Rating"])
no_keyword = np.array(no_keyword_df["Rating"])
```

```python
two_sample_test = stats.ttest_ind(mentioned_keyword, no_keyword) 

print(two_sample_test)
```

```python
keyword_mean = np.mean(mentioned_keyword) 
no_keyword_mean = np.mean(no_keyword)
print(f'Mentioned Keyword Mean: {keyword_mean}')
print(f'No Keyword Mean: {no_keyword_mean}')
```

**Conclusion:**  
On average, customers who mentioned the speed of service or checkout in their reviews had a less positive experience at the restaurant compared to customers who didnâ€™t mention these aspects in their reviews. With a two-sample t-test, we can reject the null hypothesis with a p-value of 6.19e-92 and conclude that there is a statistically significant difference in customer experiance between those who mentioned speed of service or checkout and those who didnâ€™t.

## Graph 2: Rating distribution over year

```python
df_2014 = df[df["Year"] > 2013]
keyword_df_2014 = keyword_df[keyword_df["Year"] > 2013]
no_keyword_df_2014 = no_keyword_df[no_keyword_df["Year"] > 2013]
```

```python
df_list = [df_2014, keyword_df_2014, no_keyword_df_2014]
df_name = ["Original data", "Keyword data", "non Keyword data"]
```

```python
for index in range(len(df_list)):
    grouped = df_list[index].groupby(['Year', 'Rating'])["Date"].count().reset_index()
    stacked = grouped.pivot(index="Year", columns="Rating", values="Date")
    
    fig, ax = plt.subplots(figsize=(10,5))

    stacked.plot(kind="bar", width=0.9, stacked=False, ax=ax)

    ax.set_xlabel("Year")
    ax.set_ylabel("Total Number of Reviews")
    plt.title(f"{df_name[index]}:\nTotal Number of Review by Rating over Year")
    plt.show()
```

#### - Correlation Preparation -

```python
# Remove numbers and marks, and check what/how many words in the review.
word_df = df.copy()
word_df["Review Text"] = word_df["Review Text"].str.replace(r"[^a-zA-Z]", " ").str.replace(r"\s+", " ")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk

# Let's test it out
stemmer = nltk.stem.PorterStemmer()
```

```python
# 1. Instantiate 
bagofwords = CountVectorizer(tokenizer = lambda x: [stemmer.stem(i) for i in x.split(' ')], 
                             stop_words='english')

# 2. Fit 
bagofwords.fit(word_df["Review Text"])

# 3. Transform
df_transformed = bagofwords.transform(word_df["Review Text"])
df_transformed.shape
```

```python
# All words in the reviews
review_words = bagofwords.get_feature_names()
```

```python
# Dataframe of all tokens
bagofword_df = pd.DataFrame(columns=bagofwords.get_feature_names(), data=df_transformed.toarray())
```
